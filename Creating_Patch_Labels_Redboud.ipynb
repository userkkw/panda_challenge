```
import pandas as pd
import matplotlib.pyplot as plt
import PIL
```

```
#full version
#initialize the dictionary
image_labels = {'image':[],'grade':[]}
#open the directory
directory = '/content/drive/My Drive/a51ad4ec379a9209c740025a6d410708'
#check the mask labels for each patch
for filename in os.listdir(directory):
    if filename.startswith("mask") :
        #print(os.path.join(directory, filename))
        mask_data = PIL.Image.open(filename, mode="r")
        mask_data = mask_data.split()[0]
        pixels = list(mask_data.getdata())
        replaced = filename.replace("mask", "image")
        if pixels.count(5) != 0 :
            image_labels["image"].append(replaced)
            image_labels["grade"].append(5)
        elif pixels.count(4) != 0:
            image_labels["image"].append(replaced)
            image_labels["grade"].append(4)
        elif pixels.count(3) != 0:
            image_labels["image"].append(replaced)
            image_labels["grade"].append(3)
        elif pixels.count(2) != 0:
            image_labels["image"].append(replaced)
            image_labels["grade"].append(2)
        elif pixels.count(1) != 0:
            image_labels["image"].append(replaced)
            image_labels["grade"].append(1)
    else:
        continue
#store the labels in pandas dataframe
df = pd.DataFrame(image_labels)
df
```
